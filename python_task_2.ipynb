```python
import pandas as pd

def calculate_distance_matrix(df):
    # Print column names to identify the correct columns for locations and distances
    print("Column Names:", df.columns)

    # Assign the correct column names from your dataset
    start_col = 'id_start'  # Replace with the actual column name for start locations
    end_col = 'id_end'  # Replace with the actual column name for end locations
    distance_col = 'distance'  # Replace with the actual column name for distances

    # Create a set of unique locations
    unique_locations = sorted(set(df[start_col]).union(df[end_col]))
    distance_matrix = pd.DataFrame(0, columns=unique_locations, index=unique_locations)

    # Group by start and end locations and sum distances to update the distance matrix
    for _, row in df.iterrows():
        start_loc = row[start_col]
        end_loc = row[end_col]
        distance = row[distance_col]

        # Update the distance matrix with bidirectional distances
        distance_matrix.loc[start_loc, end_loc] += distance
        distance_matrix.loc[end_loc, start_loc] += distance

    # Set diagonal values to 0
    for loc in unique_locations:
        distance_matrix.loc[loc, loc] = 0

    return distance_matrix

# Load dataset-3.csv into a DataFrame
df = pd.read_csv('dataset-3.csv')  # Replace 'dataset-3.csv' with the correct file path

# Call the function to calculate the distance matrix
distance_matrix = calculate_distance_matrix(df)
print(distance_matrix)
```

```python
import pandas as pd

# Function to calculate distance matrix
def calculate_distance_matrix(df):
    # Assuming 'df' contains columns: id_start, id_end, and distance
    unique_locations = sorted(set(df['id_start']).union(df['id_end']))
    distance_matrix = pd.DataFrame(0, columns=unique_locations, index=unique_locations)

    for _, row in df.iterrows():
        start_loc = row['id_start']
        end_loc = row['id_end']
        distance = row['distance']

        distance_matrix.loc[start_loc, end_loc] += distance
        distance_matrix.loc[end_loc, start_loc] += distance

    for loc in unique_locations:
        distance_matrix.loc[loc, loc] = 0

    return distance_matrix

# Function to unroll distance matrix
def unroll_distance_matrix(distance_matrix):
    locations = distance_matrix.index.tolist()
    id_start = []
    id_end = []
    distance = []

    for i in range(len(locations)):
        for j in range(len(locations)):
            if i != j:
                id_start.append(locations[i])
                id_end.append(locations[j])
                distance.append(distance_matrix.iloc[i, j])

    unrolled_df = pd.DataFrame({'id_start': id_start, 'id_end': id_end, 'distance': distance})

    return unrolled_df

# Function to find 'id_start' values within a 10% threshold of the reference value's average distance
def find_ids_within_ten_percentage_threshold(df, reference_value):
    reference_df = df[df['id_start'] == reference_value]
    average_distance = reference_df['distance'].mean()
    threshold_lower = average_distance - (average_distance * 0.1)
    threshold_upper = average_distance + (average_distance * 0.1)
    filtered_ids = df[(df['distance'] >= threshold_lower) & (df['distance'] <= threshold_upper)]
    sorted_filtered_ids = filtered_ids['id_start'].unique().tolist()
    sorted_filtered_ids.sort()

    return sorted_filtered_ids

# Load dataset-3.csv into a DataFrame
df = pd.read_csv('dataset-3.csv')  # Replace 'dataset-3.csv' with the correct file path

# Step 1: Calculate distance matrix
distance_matrix = calculate_distance_matrix(df)

# Step 2: Unroll distance matrix
distance_df = unroll_distance_matrix(distance_matrix)

# Step 3: Find 'id_start' values within a 10% threshold of a reference value's average distance
# Let's assume the reference value is 100 (you can replace this with your desired reference value)
reference_value = 100
result_ids = find_ids_within_ten_percentage_threshold(distance_df, reference_value)
print("IDs within 10% threshold of the reference value's average distance:", result_ids)
```
